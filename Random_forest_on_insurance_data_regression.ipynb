```python
sex = pd.get_dummies(ori_data['sex'])
```

```python
sex.head(10)
```

```python
smoker = pd.get_dummies(ori_data['smoker'])
```

```python
region = pd.get_dummies(ori_data['region'])
region
```

```python
data = pd.concat([ori_data, sex, smoker, region], axis= 1)
```

```python
data.head()
```

## Drop Original Columns

```python
data = data.drop(['sex', 'female', 'no', 'northeast'], axis=1)
```

```python
data.head(10)
```

```python
#data = data.drop('smoker', axis=1)
```

```python
data = data.drop('region', axis=1)
```

```python
X = data.drop('charges',axis=1)
y = data['charges']
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators= 50, random_state= 0)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state =1)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
